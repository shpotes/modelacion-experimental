```python
import math 

import matplotlib.pyplot as plt
import numpy as np
import sympy as sy

from src import ContinuousModel
from src.identification import transient
from src.identification.utils import compare_models
```

```python
def punto_tres(model, **kwargs):
    estim = transient.identify_second_order(
        model,
        **kwargs
    )

    print(f'estimated: {estim["equation"]}')
    
    if 'simulation_parameters' in kwargs:
        sim_kwargs = kwargs['simulation_parameters']
    else:
        sim_kwargs = {}

    compare_models({
        'real': model,
        'estimated': estim['model'],
    }, **sim_kwargs)
```

```python
# 3.a
model = ContinuousModel([2], [1, 2, 2])
punto_tres(model)
```

```python
# 3.b
model = ContinuousModel([4], [1, 1, 1], delay=2)
punto_tres(model)
```

```python
# 3.c
model = ContinuousModel([3], [1, 3, 5], delay=1)
punto_tres(model)
```
